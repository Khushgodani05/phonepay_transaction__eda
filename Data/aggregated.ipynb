```python
root_path = "../pulse/data/aggregated/insurance/country/india"
entries = os.listdir(root_path)
print(entries)
ind=0
aggregate_insurance=pd.DataFrame(columns=["name","insurancetype","insurancecount","insuranceamount","level","region","year","quater"])
for entry in entries:
    if entry != "state": 
        level = "India"
        regionlevel = "India"
        year = entry
        year_path = os.path.join(root_path, year)

        files = os.listdir(year_path)
        for file_name in files:
            try:
                file_path = os.path.join(year_path, file_name)
                with open(file_path, "r") as file:
                    response = json.load(file)

                for item in response["data"]["transactionData"]:
                    name = item["name"]
                    for instrument in item["paymentInstruments"]:
                        type = instrument["type"]
                        count = instrument["count"]
                        amount = instrument["amount"]
                        quarter = file_name.split(".")[0]
                        aggregate_insurance.loc[ind]=[name,type,count,amount,level,regionlevel,year,quarter]
                        ind+=1

            except Exception as e:
                print("Country error:", file_path, e)

    else:  
        level = "state"
        state_root = os.path.join(root_path, "state")
        states = os.listdir(state_root)

        for state_name in states:
            state_path = os.path.join(state_root, state_name)
            regionlevel=state_name
            if os.path.isdir(state_path):
                for year in os.listdir(state_path):
                    year_path = os.path.join(state_path, year)

                    if os.path.isdir(year_path):
                        for file_name in os.listdir(year_path):
                            try:
                                file_path = os.path.join(year_path, file_name)
                                with open(file_path, "r") as file:
                                    response = json.load(file)

                                for item in response["data"]["transactionData"]:
                                    name = item["name"]
                                    for instrument in item["paymentInstruments"]:
                                        type = instrument["type"]
                                        count = instrument["count"]
                                        amount = instrument["amount"]
                                        quarter = file_name.split(".")[0]
                                        aggregate_insurance.loc[ind]=[name,type,count,amount,level,regionlevel,year,quarter]
                                        ind+=1

                            except Exception as e:
                                print("State error:", file_path, e)
                                
aggregate_insurance
```

### Aggregated users 

```python
root_path="../pulse/data/aggregated/user/country/india"
entries=os.listdir(root_path)
aggregate_users=pd.DataFrame(columns=["appopen","usersbydevicebrand","usersbydevicecount","level","region","year","quater"])
ind=0
print(entries)
for entry in entries:
    if entry!="state":
        level="India"
        year=entry
        regionlevel="India"
        year_path=os.path.join(root_path,entry)
        files=os.listdir(year_path)
        for filename in files:
            try:
                filepath=os.path.join(year_path,filename)
                with open(filepath,"r") as f:
                    response=json.load(f)
                    obj=response["data"]
                    appopen=obj["aggregated"]["appOpens"]
                    quarter=filename.split(".")[0]
                    deviceitem=obj["usersByDevice"]
                    if deviceitem:
                        for user in deviceitem:
                            brand=user["brand"]
                            count=user["count"]
                            aggregate_users.loc[ind]=[appopen,brand,count,level,regionlevel,year,quarter]
                            ind+=1
                    else:
                        brand="Nan"
                        count="Nan"
                        aggregate_users.loc[ind]=[appopen,brand,count,level,regionlevel,year,quarter]
                        ind+=1
            except Exception as e:
                print(e)
    else:
        level="state"
        state_path=os.path.join(root_path,entry)
        states=os.listdir(state_path)
        for i in states:
            regionlevel= i
            stateinfopath=os.path.join(state_path,i)
            if os.listdir(stateinfopath):
                years=os.listdir(stateinfopath)
                for yr in years:
                    year=yr
                    year_path=os.path.join(stateinfopath,yr)
                    datas=os.listdir(year_path)
                    for files in datas:
                        try:
                            filepath=os.path.join(year_path,files)
                            with open(filepath,"r") as file:
                                response=json.load(file)
                                obj=response["data"]
                                appopen=obj["aggregated"]["appOpens"]
                                quarter=files.split(".")[0]
                                deviceitem=obj["usersByDevice"]
                                if deviceitem:
                                    for user in deviceitem:
                                        brand=user["brand"]
                                        count=user["count"]
                                        aggregate_users.loc[ind]=[appopen,brand,count,level,regionlevel,year,quarter]
                                        ind+=1
                                else:
                                    brand="Nan"
                                    count="Nan"
                                    aggregate_users.loc[ind]=[appopen,brand,count,level,regionlevel,year,quarter]
                                    ind+=1
                        except Exception as e:
                            print(e)
```
